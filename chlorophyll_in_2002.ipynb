```
import matplotlib.pyplot as plt #allows us to visualize data
import numpy as np #allows us to use mathematical functions
import csv #csv allows us to read values separated by a comma
```

```
csvFile = open("/content/drive/My Drive/chlorophyll_mgm3_Dec_2002.csv")
csvReader = csv.reader(csvFile, delimiter=',') 
data_lists = list(csvReader)
```

```
print(data_lists[0])
```

```
print(len(data_lists[0]))
print(len(data_lists))
```

```
width = 720
height = 360
```

```
grid = np.empty([height, width, 3], dtype=np.uint8)
#creates an empty matrix which is a grid made up of lists
#also tells code to take in three number values (for the RGB values)
```

```
all_values = []

for r, row in enumerate(data_lists): #keeps track of the value and the index making it easier for the program to keep track of the nested loop
  for c, value in enumerate(row):
    if float(value) < 99999:
      all_values.append(float(value))
    if float(value) <= 0.0111:
      grid[r, c] = [55, 50, 0]
    elif float(value) <= 0.02:
      grid[r, c] = [102, 0, 0]
    elif float(value) <= 0.03:
      grid[r, c] = [153, 0, 0]
    elif float(value) <= 0.04:
      grid[r, c] = [204, 0, 0]
    elif float(value) <= 0.05:
      grid[r, c] = [255, 0, 0]    
    elif float(value) <= 0.07:
      grid[r, c] = [255, 128, 0]    
    elif float(value) <= 0.1:
      grid[r, c] = [255, 248, 31]       
    elif float(value) <= 0.135:
      grid[r, c] = [0, 255, 221]
    elif float(value) <= 0.15:
      grid[r, c] = [0, 195, 255]
    elif float(value) <= 0.2:
      grid[r, c] = [0, 50, 255]    
    elif float(value) <= 0.3:
      grid[r, c] = [0, 0, 204]    
    elif float(value) <= 0.4:
      grid[r, c] = [0, 0, 102]    
    elif float(value) <= 0.5:
      grid[r, c] = [123, 5, 250]    
    elif float(value) <= 0.6:
      grid[r, c] = [212, 0, 255]    
    elif float(value) <= 0.9:
      grid[r, c] = [166, 2, 247]    
    elif float(value) <= 10:
      grid[r, c] = [250, 5, 250]    
    elif float(value) <= 30:
      grid[r, c] = [201, 0, 144]       
    elif float(value) <= 69.2596:
      grid[r, c] = [255, 255, 255]    
    else:
      grid[r, c] = [93, 97, 99]
```

```
print(min(all_values))
print(max(all_values))
```

```

plt.title("Amount of Chlorophyll in the Sea (in 2002)")
plt.imshow(grid)
plt.axis('off')
plt.savefig("Land_temperature.png",dpi = 1000)
plt.ylim(340, 30)
plt.show()
```
